```python
name = input("Enter your name: ") # Prompt user for their name
age = int(input("Enter your age: ")) # Prompt user for their age and convert to integer
city = input("Enter your city: ") # Prompt user for their city
job_title = input("What job do you so? ") # Prompt user for their job title
salary = int(input("What is your salary?")) # Prompt user for their salary and convert to integer
favorite_skill = input("What is your favorite skill: ") # Prompt user for their favorite skill

print("\n--- User Information ---") # Print a header for user information
# Display the collected user information in a formatted way
print(f"{"Name:":<20} {name}")
print(f"{"Age:":<20} {age}")
print(f"{"City:":<20} {city}")
print(f"{"Job Title:":<20} {job_title}")
print(f"{"Salary:":<20} {salary:,}")
print(f"{"Favorite Skill:":<20} {favorite_skill}")
print("------------------------") # Print a footer
```
